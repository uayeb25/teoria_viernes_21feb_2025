```python
def extend_one_numeric_column_one_cat( df , cat , num , size ):
    cats = df.loc[ : , [cat] ].drop_duplicates()[cat].tolist()

    fi =  df.groupby( cat, as_index=False ).agg({
        num: ["count"]
    })
    fi.columns = [ cat , "n" ]
    fi["n_prod"] = fi["n"] / df.shape[0]


    return pd.DataFrame({
        cat: np.random.choice( cats , size , p=fi["n_prod"].tolist() )
        , num: [ random.randint( np.min( df[num] ) , np.max( df[num] ) ) for _ in range(size) ] 
    })
```

```python
extend_one_numeric_column_one_cat( df , "category" , "protein" , 2000 )
```
