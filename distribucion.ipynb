```python
import pandas as pd
import numpy as np
import random

import matplotlib.pyplot as plt
from scipy.stats import truncnorm
```

```python
df = pd.DataFrame({
    "num": [ random.randint(0,100) for _ in range(20000) ]
})
```

```python
df.describe()
```

```python
df.hist()
```

```python
df = pd.DataFrame({
    "num": [ random.gauss(0,100) for _ in range(20000) ]
})
```

```python
df.hist()
```

```python
df = pd.DataFrame({
    "num": [ random.expovariate(7) for _ in range(2000) ]
})
```

```python
df.hist()
```

```python
min_val = 1
max_val = 100

media = 80
desv_std = 5

a, b = ( min_val - media ) / desv_std, ( max_val - media) / desv_std
```

```python
muestras = truncnorm.rvs( a , b , loc=media, scale=desv_std, size=20000)
```

```python
muestras = np.append(muestras, 1)
```

```python
df = pd.DataFrame({
    "num": muestras
})
```

```python
df.describe()
```

```python
df.hist()
```

```python
df.boxplot()
```
